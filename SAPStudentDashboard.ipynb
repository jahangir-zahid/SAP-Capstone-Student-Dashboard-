```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# student data
data = {
    "Abdul Basit" : {
        "Maths" : 44,
        "Physics" : 40,
        "AI" : 50,
        "Java" : 45
    },
    "Jahangir Ali" : {
        "Maths" : 50,
        "Physics" : 40,
        "AI" : 45,
        "Java" : 53,
    },
    "Mohammad Ghouse Khan" : {
        "Maths" : 40,
        "Physics" : 35,
        "AI" : 42,
        "Java" : 39
    },
    "Syed Zumair Sohail" : {
        "Maths" : 49,
        "Physics" : 46,
        "AI" : 52,
        "Java" : 47,
    }
}

df = pd.DataFrame(data)

#Display all students
def display_students():
    print("\n--- Student Records ---")
    print(df)
    print("="*80)

#Student result
def student_result():
    name = input("Enter student name: ").title()
    if name in data:
        print(f"\nStudent Name: {name}")
        for subject, marks in data[name].items():
            print(subject, ":", marks)
    else:
        print("Student not found")
    print("="*80)

#Class Average
def class_average():
    avg = ["Maths", "Physics", "AI", "Java"]
    avg = df.mean(axis=1)
    print("\n--- Class Average Marks ---")
    print(avg)
    print("="*80)


#Charts For All Students Average Marks
def plot_average_marks():
    avg = df.mean(axis=1)
    # Bar Chart
    plt.figure()
    avg.plot(kind="bar")
    plt.title("Class Average Marks (Bar Chart)")
    plt.xlabel("Subjects")
    plt.ylabel("Average Marks")
    plt.show()
    # Pie Chart
    plt.figure()
    plt.pie(
        avg,
        labels=avg.index,
        autopct='%1.1f%%',
        startangle=90
    )
    plt.title("Class Average Marks (Pie Chart)")
    plt.show()
    print("="*80)


#Charts for Seprate students marks 
def all_students_charts():
    for name, subjects in data.items():
        subject_names = list(subjects.keys())
        marks = list(subjects.values())
        # BAR GRAPH
        plt.figure()
        plt.bar(subject_names, marks)
        plt.title(f"{name} - Marks (Bar Graph)")
        plt.xlabel("Subjects")
        plt.ylabel("Marks")
        plt.ylim(0, 60)
        plt.show()
        # PIE CHART
        plt.figure()
        plt.pie(
            marks,
            labels=subject_names,
            autopct='%1.1f%%',
            startangle=90
        )
        plt.title(f"{name} - Marks Distribution (Pie Chart)")
        plt.show()
    print("="*80)


#improvement in subjects
def improvement_suggestion():
    print("\n--- Automatic Improvement Suggestion Report ---")
    threshold = 45 
    for subject in df.index:
        subject_marks = df.loc[subject]
        avg = subject_marks.mean()
        print(f"\nSubject: {subject}")
        print(f"Average Marks: {avg:.2f}")
        if avg < threshold:
            print("Improvement Required in this subject")
        else:
            print("Overall performance is good")
        print("Students needing improvement:")
        need_help = False
        for student, marks in subject_marks.items():
            if marks < threshold:
                print(f"- {student} ({marks})")
                need_help = True
        if not need_help:
            print("None")
    print("="*80)

#Looping statement
while True:
    print("\n========== Student Dashboard ==========")
    print("1. View All Students")
    print("2. View Student Result")
    print("3. View Class Average")
    print("4. Show Average Marks Chart")
    print("5. Show Charts for All Students")
    print("6. Show Improvement Suggestions")
    print("7. Exit")

    try:
        choice = int(input("Enter your choice: "))
        match choice:
            case 1:
                display_students()
            case 2:
                student_result()
            case 3:
                class_average()
            case 4:
                plot_average_marks()
            case 5:
                all_students_charts()
            case 6:
                improvement_suggestion()
            case 7:
                print("\nExiting Student Dashboard. Thank you!")
                break
            case _:
                print("Invalid choice! Please select 1 - 7.")
    except ValueError:
        print("Please enter a valid number.")
```
